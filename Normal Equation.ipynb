##  **Normal equation**

### **Trick!!!!**
-  **.c_** ใช้ merge 2 metrix เข้าด้วยกัน
-  **np.linalg.inv** เป็น function ของ library numpy ใช้ inverse
    - กรณีเป็น singular metrix = det = 0  function **.inv หาค่าไม่ได้**  
      **แก้ไขโดย**  ใช้ Moore-Pennes โดย Function **.pinv** ทำให้ Automatic

```python
import numpy as np
x = np.array([0,2,3])
y = np.array([1,1,4])
x_b = np.c_[np.ones((3,1)),x]
# print(x_b)
# x_b = [[1. 0.]
#       [1. 2.]
#       [1. 3.]]


#call inverse function from linear algebra module
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
print("theta = ", theta_best)

#prediction
x_n = np.array([[0],[10]])
x_n_b = np.c_[np.ones((2,1)), x_n]
print(x_n_b)
y_p = x_n_b.dot(theta_best)
print("y predict = ",y_p)

import matplotlib.pyplot as plt
plt.plot(x_n, y_p, "r-")
plt.plot(x,y, "b.")
plt.axis([0,10,0,10])
plt.show()
```

##  ข้อนี้กรณีเป็น singular metrix = det = 0  function <.inv>  หาค่าไม่ได้
## **แก้ไขโดย**  ใช้ Moore-Penroes-inverse โดย Function <.pinv> ทำให้ Automatic

```python
import numpy as np
x = np.array([2,2,2])
y = np.array([1,1,4])
x_b = np.c_[np.ones((3,1)), x]

print(x)
print(y)

#call inverse function from linear algebra module
theta_best = np.linalg.pinv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)   # เปลี่ยน inv เป็น pinv เพื่อหาค่า moore penrose กรณี det = 0 singular matrix
print("theta = ", theta_best)

#prediction
x_n = np.array([[0],[10]])
print(x_n)
x_n_b = np.c_[np.ones((2,1)), x_n]
print(x_n_b)
y_p = x_n_b.dot(theta_best)
print("y predict = ",y_p)

import matplotlib.pyplot as plt
plt.plot(x_n, y_p,"r-")
plt.plot(x,y, "b.")
plt.axis([0,10,0,10])
plt.show()
```

### Sklearn

```python
import numpy as np
from sklearn.linear_model import LinearRegression 
x = np.array([[0,2,3]]).T
y = np.array([1,1,4])

lin_reg = LinearRegression()
lin_reg.fit(x, y)
print(lin_reg.intercept_, " , ", lin_reg.coef_)
x_n = np.array([[0]]).T
ans = lin_reg.predict(x_n)
print(ans)
```
